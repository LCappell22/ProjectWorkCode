```python
import matplotlib.pyplot as plt
import scipy.constants as sc
import numpy as np

#importing data from file
infilename = 'M1profile5.data' #limited set of vars, easy to parse
file2 = 'M0.3profile5.data'
data = np.genfromtxt(infilename, unpack=True)
data2 = np.genfromtxt(file2, unpack=True)

#extracting necessary data from the files
M1radius = data[1]
M1rho = (10**data[4]) #files have rho in log, so must raise to the power of 10 to properly obtain density

M03radius = data2[1]
M03rho = (10**data2[4])
```

```python
#plots extracted data for 1 solar mass star

fig = plt.figure()
fig.suptitle("Density against radius of M = 1M☉")
plt.plot(M1radius, M1rho)
ax = plt.gca()
ax.grid()
ax.set_xlabel("Radius ($R☉$)")
ax.set_ylabel("Density ($g$ $cm^{-3}$)")
plt.show()
```

```python
#plots extracted data for 0.3 solar mass star

fig = plt.figure()
fig.suptitle("Density against radius of M = 0.3M☉")
plt.plot(M03radius, M03rho)
ax = plt.gca()
ax.grid()
ax.set_xlabel("Radius ($R☉$)")
ax.set_ylabel("Density ($g$ $cm^{-3}$)")
plt.show()
```
